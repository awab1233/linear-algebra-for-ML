- debugging decompositions

Exercise 2-10.
An important skill in coding is finding bugs. Let’s say there is a bug in your code
such that the denominator in the projection scalar of Equation 2-13 is tT
t instead of
rT
r (an easy mistake to make, speaking from personal experience while writing this
chapter!). Implement this bug to check whether it really deviates from the accurate
code. What can you do to check whether the result is correct or incorrect? (In coding,
confirming code with known results is called sanity-checking.)

$$\beta = \frac{t \cdot r}{r \cdot r} = \frac{t^T r}{\|r\|^2}$$

$$t_{\parallel r} = \beta \times r$$$$t_{\perp r} = t - t_{\parallel r}$$

```python
t = np.random.randn(2)
r = np.random.randn(2)

# حساب المعامل (Scalar)
beta = (np.dot(t,r) / np.dot(r,r))

# المكون الموازي (Parallel Component)
para_t = beta * r

# المكون العمودي (Perpendicular Component)
perp_t = t - para_t


is_orthogonal = np.isclose(np.dot(para_t, perp_t), 0, atol=1e-15)
print(int(is_orthogonal))
print(f"Is it truly orthogonal? {is_orthogonal}")
```
